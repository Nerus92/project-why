# Dynamic Array Exercise

```python
import sys, ctypes


class DynamicArray(object):
    
    def __init__(self):
        self.elements_count = 0
        self.capacity = 1
        self.A = self._make_array(self.capacity)
        
    def __len__(self):
        return self.elements_count
    
    def __getitem__(self, index):
        if not 0 <= index < self.elements_count:
            return IndexError('%d is out of bounds' % index)
        else:
            return self.A[index]
        
    def append(self, item):
        if self.elements_count == self.capacity:
            self._resize(2 * self.capacity)
        self.A[self.elements_count] = item
        self.elements_count += 1
            
    def _resize(self, new_capacity):
        B = self._make_array(new_capacity)
        for index in range(self.elements_count):
            B[index] = self.A[index]
        self.capacity = new_capacity
        self.A = B
        
    def _make_array(self, size):
        return (size * ctypes.py_object)()
    
    def __str__(self):
        return ', '.join(str(self.A[index]) for index in range(self.elements_count))
```

```python
dyn = DynamicArray()
for i in range(100):
    dyn.append(i)
    print(ctypes.sizeof(dyn.A))
print('DYN: %s' % dyn)
```
